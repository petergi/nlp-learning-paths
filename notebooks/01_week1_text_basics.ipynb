# Week 1: Text Basics

Goal: clean text, tokenize, and remove stopwords.

```python
from nlp_learning.preprocessing import clean_text, tokenize, remove_stopwords

sample = "NLP is amazing! It helps computers understand human language."
print("Original:", sample)
print("Cleaned:", clean_text(sample))
```

```python
tokens = tokenize(sample)
tokens_no_stopwords = remove_stopwords(tokens)

print("Tokens:", tokens)
print("Without stopwords:", tokens_no_stopwords)
```
