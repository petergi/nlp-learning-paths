# Week 2: Bag-of-Words and TF-IDF

Goal: represent text as numeric features.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

df = pd.read_csv('../data/sample_reviews.csv')
df
```

```python
count_vec = CountVectorizer()
X_count = count_vec.fit_transform(df['text'])

print('Count matrix shape:', X_count.shape)
print('Sample vocabulary:', list(count_vec.vocabulary_.keys())[:10])
```

```python
tfidf_vec = TfidfVectorizer()
X_tfidf = tfidf_vec.fit_transform(df['text'])

print('TF-IDF matrix shape:', X_tfidf.shape)
print('A TF-IDF row as dense values:', X_tfidf[0].toarray()[0][:10])
```
