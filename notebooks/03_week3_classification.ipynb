# Week 3: Supervised Text Classification

Goal: train and evaluate a classifier on labeled text data.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

df = pd.read_csv('../data/sample_reviews.csv')
df['target'] = (df['label'] == 'positive').astype(int)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['target'], test_size=0.4, random_state=42, stratify=df['target']
)

vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = LogisticRegression(max_iter=500)
model.fit(X_train_vec, y_train)
preds = model.predict(X_test_vec)

print(classification_report(y_test, preds, target_names=['negative', 'positive']))
```

```python
examples = ['I loved this experience', 'This was awful and boring']
examples_vec = vectorizer.transform(examples)
pred_labels = model.predict(examples_vec)

for text, label in zip(examples, pred_labels):
    print(text, '->', 'positive' if label == 1 else 'negative')
```
