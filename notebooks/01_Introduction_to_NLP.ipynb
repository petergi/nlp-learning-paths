## 1.1 What is Natural Language Processing (NLP)?

**Natural Language Processing (NLP)** is a subfield of artificial intelligence that focuses on the interaction between computers and humans through natural language. The ultimate goal of NLP is to enable computers to **understand, interpret, and generate** human languages in a way that is both meaningful and useful.

NLP blends **linguistics, computer science, and artificial intelligence**. It enables machines to interpret, understand, and respond to human language in a valuable way.

### Key Areas of NLP

NLP encompasses a wide range of techniques that can be broadly classified into:

| Category | Description | Examples |
|---|---|---|
| **Text Processing** | Breaking down and cleaning text | Tokenization, stemming, lemmatization, stop word removal |
| **Syntactic Analysis** | Understanding grammatical structure | Part-of-speech tagging, parsing |
| **Semantic Analysis** | Understanding meaning | Word sense disambiguation, semantic role labeling |
| **Pragmatic Analysis** | Understanding context and intent | Sarcasm detection, dialogue systems |

### 1.1.1 Definition and Scope of NLP

NLP covers a broad range of capabilities:

- **Text Processing**: Tokenization (breaking text into words/tokens), stemming (reducing words to root form), lemmatization (converting to dictionary form), and text cleaning.
- **Syntactic Analysis**: Parsing sentences to understand grammatical structure, identifying parts of speech and their relationships.
- **Semantic Analysis**: Understanding word meaning in context -- word sense disambiguation and semantic role labeling.
- **Pragmatic Analysis**: Going beyond literal meaning to understand speaker intent, situational context, and cultural factors.

### 1.1.2 Applications of NLP

NLP powers a wide range of real-world technologies:

- **Search Engines**: Google uses NLP to understand user queries and return relevant results
- **Machine Translation**: Google Translate converts text between languages while preserving meaning
- **Chatbots & Virtual Assistants**: Siri, Alexa, and Google Assistant understand natural language commands
- **Sentiment Analysis**: Businesses analyze customer feedback to gauge public opinion
- **Text Summarization**: Automatic generation of concise summaries from long documents
- **Spam Detection**: Email services filter spam by analyzing content and context
- **Speech Recognition**: Converting spoken language into written text
- **Recommendation Systems**: Netflix and Amazon analyze reviews to recommend content
- **Healthcare**: Analyzing patient records and clinical notes for better care
- **Legal Tech**: Reviewing legal documents and contracts for key information

### 1.1.3 Importance of NLP

NLP is a transformative technology because it:

1. **Bridges communication gaps** between humans and machines
2. **Automates repetitive tasks** like email sorting, spam filtering, and customer service
3. **Improves accessibility** through speech recognition and text-to-speech for individuals with disabilities
4. **Enables large-scale data analysis** of unstructured text data
5. **Powers personalization** in e-commerce and streaming platforms

### Challenges in NLP

Despite its power, NLP faces significant challenges:
- **Ambiguity**: "bank" can mean a financial institution or a river bank
- **Context dependence**: "bat" means different things in different contexts
- **Language diversity**: Models trained on English may not work for Chinese or Arabic
- **Idioms**: "kick the bucket" cannot be interpreted literally
- **Sarcasm/Irony**: "Oh great, another traffic jam!" is negative, not positive
- **Domain-specific knowledge**: Medical text differs vastly from legal text

### 1.1.4 Example: Tokenization in NLP

**Tokenization** is the process of breaking down text into smaller units called **tokens** (words, phrases, or characters). It is a fundamental first step in nearly all NLP tasks.

````python
import nltk
from nltk.tokenize import word_tokenize

# Sample text
text = "Natural Language Processing (NLP) enables machines to understand human language."

# Tokenize the text
tokens = word_tokenize(text)

# Display the tokens
print("Tokens:", tokens)
print(f"\nNumber of tokens: {len(tokens)}")
````

**Why tokenization matters:**
- It is the first step in many NLP tasks (text analysis, machine translation, sentiment analysis)
- It prepares data for further analysis like frequency distribution and part-of-speech tagging
- It simplifies complex text into manageable pieces for algorithms to process

Notice how punctuation like `(`, `)`, and `.` are treated as separate tokens.

### 1.2.2 Sentiment Analysis Example

Sentiment analysis uses NLP to determine whether text expresses **positive**, **negative**, or **neutral** sentiment. This is invaluable for businesses analyzing customer feedback.

The `TextBlob` library provides a simple interface for sentiment analysis, returning:
- **Polarity**: ranges from -1 (very negative) to +1 (very positive)
- **Subjectivity**: ranges from 0 (objective/fact-based) to 1 (subjective/opinion-based)

````python
from textblob import TextBlob

# Sample text
text = "I love this product! It's amazing."

# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment
print(f"Text: '{text}'")
print(f"Polarity:     {sentiment.polarity:.2f}  (range: -1 to 1)")
print(f"Subjectivity: {sentiment.subjectivity:.2f}  (range: 0 to 1)")
````

### 1.2.3 Text Summarization Example

NLP techniques can automatically create concise summaries of lengthy documents. The `sumy` library implements several summarization algorithms, including **Latent Semantic Analysis (LSA)**.

````python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

# Sample text
text = """
Natural Language Processing (NLP) is a fascinating field at the intersection of
computer science, artificial intelligence, and linguistics. It enables machines to
understand, interpret, and generate human language, opening up a world of possibilities
for applications ranging from chatbots and translation services to sentiment analysis
and beyond.
"""

# Create a parser
parser = PlaintextParser.from_string(text, Tokenizer("english"))

# Create a summarizer using LSA
summarizer = LsaSummarizer()

# Generate the summary (2 sentences)
summary = summarizer(parser.document, 2)

print("Summary:")
for sentence in summary:
    print(f"  - {sentence}")
````

### 1.2.4 Real-World Example: E-commerce Review Analysis

A practical application of NLP in e-commerce is analyzing customer reviews to extract insights. Here we use NLTK's **VADER** (Valence Aware Dictionary and sEntiment Reasoner) sentiment analyzer, which is specifically designed for social media and short texts.

VADER returns four scores:
- `neg`: Negative sentiment proportion
- `neu`: Neutral sentiment proportion
- `pos`: Positive sentiment proportion
- `compound`: Normalized overall score (-1 to +1)

````python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Sample e-commerce reviews
reviews = [
    "This product is fantastic! It exceeded my expectations.",
    "Not worth the price. I'm disappointed with the quality.",
    "Good value for money. Will buy again.",
]

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Analyze each review
for review in reviews:
    sentiment = sia.polarity_scores(review)
    
    # Determine overall sentiment label
    compound = sentiment['compound']
    if compound >= 0.05:
        label = "POSITIVE"
    elif compound <= -0.05:
        label = "NEGATIVE"
    else:
        label = "NEUTRAL"
    
    print(f"Review: {review}")
    print(f"  Scores: {sentiment}")
    print(f"  Overall: {label}")
    print()
````

### 1.3.2 spaCy: Named Entity Recognition (NER)

spaCy is a modern, efficient library designed for production-level NLP. It excels at tasks like tokenization, NER, and POS tagging with pre-trained models.

````python
import spacy

# Load the small English model
nlp = spacy.load("en_core_web_sm")

text = "Apple is looking at buying U.K. startup for 1 billion."
doc = nlp(text)

# Extract named entities
print(f"Text: '{text}'\n")
print(f"{'Entity':<15} {'Label':<10} {'Description'}")
print("-" * 50)
for ent in doc.ents:
    print(f"{ent.text:<15} {ent.label_:<10} {spacy.explain(ent.label_)}")
````

spaCy identifies:
- **Apple** as an ORG (Organization)
- **U.K.** as a GPE (Geopolitical Entity)
- **1 billion** as MONEY

### 1.3.3 gensim: Word2Vec Word Embeddings

gensim specializes in topic modeling and word embeddings. **Word2Vec** transforms words into dense vector representations that capture semantic relationships.

````python
from gensim.models import Word2Vec

# Sample sentences (each sentence is a list of words)
sentences = [
    ["natural", "language", "processing"],
    ["python", "is", "a", "powerful", "language"],
    ["text", "processing", "with", "gensim"],
]

# Train a Word2Vec model
model = Word2Vec(
    sentences,
    vector_size=100,  # Dimensionality of word vectors
    window=5,          # Context window size
    min_count=1,       # Minimum word frequency
    workers=4          # Number of worker threads
)

# Get the vector for the word 'language'
vector = model.wv['language']
print(f"Vector for 'language' (first 10 dimensions):")
print(vector[:10])
print(f"\nFull vector dimensionality: {len(vector)}")
print(f"\nVocabulary: {list(model.wv.index_to_key)}")
````

Each word is represented as a 100-dimensional vector. Words with similar meanings will have vectors that are close to each other in the vector space. Applications include text classification, clustering, and recommendation systems.

### 1.3.4 scikit-learn: Text Classification

scikit-learn provides tools for building machine learning models. Here we use **CountVectorizer** for feature extraction and **Naive Bayes** for classification.

````python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample data: texts and their sentiment labels (1=positive, 0=negative)
texts = [
    "I love this product",
    "This is the worst experience",
    "Absolutely fantastic!",
    "Not good at all"
]
labels = [1, 0, 1, 0]

# Step 1: Convert text to numerical features (bag of words)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

print("Feature names:", vectorizer.get_feature_names_out())
print("Feature matrix shape:", X.shape)
print()

# Step 2: Train a Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X, labels)

# Step 3: Predict sentiment for new text
new_text = ["I hate this"]
X_new = vectorizer.transform(new_text)
prediction = classifier.predict(X_new)
print(f"Text: '{new_text[0]}'")
print(f"Predicted sentiment: {'Positive' if prediction[0] == 1 else 'Negative'}")
````

This demonstrates the fundamental text classification pipeline:
1. **Data Preparation** -- collect and label text data
2. **Feature Extraction** -- convert text to numerical features with CountVectorizer
3. **Model Training** -- train a classifier on the vectorized features
4. **Prediction** -- classify new, unseen text

## 1.4 End-to-End NLP Pipeline Example

This example combines multiple NLP tools into a complete sentiment analysis pipeline:
1. **Custom tokenizer** using spaCy
2. **Stop word removal** using NLTK
3. **Feature extraction** using CountVectorizer
4. **Classification** using Naive Bayes

The `sklearn.pipeline.Pipeline` chains these steps together for streamlined training and prediction.

````python
import nltk
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from nltk.corpus import stopwords

# Sample data
texts = [
    "I love this product! It's amazing.",
    "This is the worst experience I've ever had.",
    "Absolutely fantastic! Highly recommend.",
    "Not good at all. Very disappointing."
]
labels = [1, 0, 1, 0]  # 1 = positive, 0 = negative

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Custom tokenizer using spaCy
def spacy_tokenizer(sentence):
    doc = nlp(sentence)
    return [token.text for token in doc]

# Get English stop words
stop_words = set(stopwords.words('english'))

# Define the pipeline: vectorizer -> classifier
pipeline = Pipeline([
    ('vectorizer', CountVectorizer(tokenizer=spacy_tokenizer, stop_words=stop_words)),
    ('classifier', MultinomialNB())
])

# Train the model
pipeline.fit(texts, labels)

# Predict sentiment for new texts
test_texts = ["I hate this product", "What a wonderful experience!", "Terrible quality"]
predictions = pipeline.predict(test_texts)

for text, pred in zip(test_texts, predictions):
    sentiment = "Positive" if pred == 1 else "Negative"
    print(f"'{text}' -> {sentiment}")
````

**Pipeline summary:**
1. **Data Preparation**: Labeled text data (positive/negative)
2. **Custom Tokenizer**: spaCy breaks text into tokens
3. **Stop Word Removal**: Common English words are filtered out
4. **Feature Extraction**: CountVectorizer converts tokens to numerical features
5. **Model Training**: Naive Bayes learns to associate features with sentiment labels
6. **Prediction**: New text is processed through the same pipeline and classified

### Exercise 5: Text Classification with scikit-learn

**Task:** Use scikit-learn to train a Naive Bayes classifier on the data below, then predict the sentiment of: **"This experience was fantastic."**
