## 2.1 Understanding Text Data

Text data is inherently unstructured and can come in various forms: articles, social media posts, emails, chat messages, reviews, and more. Unlike numerical data, text requires special handling and processing techniques to convert it into a structured format that algorithms can work with.

### 2.1.1 Nature of Text Data

Text data consists of sequences of characters forming words, sentences, and paragraphs. Each piece of text can vary greatly in length, structure, and content. This variability poses challenges for analysis, as the text must be standardized and cleaned before any meaningful processing can occur.

### 2.1.2 Why Preprocessing Matters

- **Noise Reduction**: Removing irrelevant information (punctuation, stop words, special characters)
- **Standardization**: Converting text to a uniform format (lowercasing, stemming, lemmatization)
- **Feature Extraction**: Transforming text into features (tokenization, vectorization, embeddings)

### 2.1.3 Example: Exploring Raw Text Data

Let's start by exploring raw text data using Python to examine its basic properties.

```python
# Sample text
text = "Natural Language Processing (NLP) enables computers to understand human language."

# Display the text
print("Original Text:")
print(text)

# Length of the text
print("\nLength of the text:", len(text))

# Unique characters in the text
unique_characters = set(text)
print("\nUnique characters:", unique_characters)

# Number of words in the text
words = text.split()
print("\nNumber of words:", len(words))

# Display the words
print("\nWords in the text:")
print(words)
```

**Key observations:**
- `len(text)` counts all characters including spaces and punctuation
- `set(text)` gives us every unique character
- `text.split()` performs a naive word split on whitespace

Notice that `split()` keeps punctuation attached to words (e.g., `'language.'` and `'(NLP)'`). This is why we need proper tokenization and cleaning.

### 2.1.4 Challenges with Text Data

| Challenge | Description | Example |
|---|---|---|
| **Ambiguity** | Words can have multiple meanings depending on context | "bank" (river bank vs. financial institution) |
| **Variability** | Different formats, styles, and structures across sources | Formal articles vs. informal tweets |
| **Noisy Data** | Irrelevant information like punctuation, HTML tags, stop words | `<p>Hello!</p>` |
| **High Dimensionality** | Large vocabularies create very high-dimensional feature spaces | Each unique word = one dimension |
| **Sarcasm & Irony** | Tone-dependent expressions are hard for algorithms to detect | "Oh great, another meeting" |

### 2.1.5 Practical Example: Basic Text Preprocessing Steps

A basic preprocessing pipeline: lowercasing, removing punctuation, and tokenization.

```python
import string

# Sample text
text = "Natural Language Processing (NLP) enables computers to understand human language."

# Step 1: Convert to lowercase
text = text.lower()
print("Lowercased Text:")
print(text)

# Step 2: Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))
print("\nText without Punctuation:")
print(text)

# Step 3: Tokenize the text
tokens = text.split()
print("\nTokens:")
print(tokens)
```

## 2.2 Text Cleaning: Stop Word Removal, Stemming, Lemmatization

Text cleaning transforms raw, messy text into a clean and standardized format. In this section we cover three essential techniques:

1. **Stop Word Removal** -- Eliminating common words that add little semantic value
2. **Stemming** -- Reducing words to their root form by chopping off affixes
3. **Lemmatization** -- Reducing words to their dictionary (lemma) form with context awareness

### 2.2.1 Stop Word Removal

Stop words are common words like "the", "is", "in", "and" that appear frequently but carry minimal meaningful information. Removing them helps:
- **Reduce dimensionality** of the feature space
- **Speed up** processing
- **Improve accuracy** by letting algorithms focus on meaningful terms

```python
from nltk.corpus import stopwords

# Sample text
text = "Natural Language Processing enables computers to understand human language."

# Tokenize the text
tokens = text.split()

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Original Tokens:")
print(tokens)
print("\nFiltered Tokens (stop words removed):")
print(filtered_tokens)
print(f"\nRemoved {len(tokens) - len(filtered_tokens)} stop words")
```

```python
# Let's see what English stop words look like
print("Total English stop words:", len(stop_words))
print("\nFirst 20 stop words (sorted):")
print(sorted(stop_words)[:20])
```

### 2.2.2 Stemming

Stemming reduces words to their base or root form by removing suffixes, prefixes, or other affixes. For example, "running", "runner", and "runs" all reduce to "run".

The most commonly used stemming algorithm is the **Porter Stemmer**.

**Limitations:**
- **Overstemming**: Too aggressive (e.g., "university" -> "univers")
- **Understemming**: Not aggressive enough (e.g., "organization" and "organizing" may not share a stem)
- **Context ignorance**: Does not consider word meaning

```python
from nltk.stem import PorterStemmer

# Sample text
text = "Natural Language Processing enables computers to understand human language."

# Tokenize the text
tokens = text.split()

# Initialize the stemmer
stemmer = PorterStemmer()

# Stem the tokens
stemmed_tokens = [stemmer.stem(word) for word in tokens]

print("Original Tokens:")
print(tokens)
print("\nStemmed Tokens:")
print(stemmed_tokens)
```

```python
# Demonstrate stemming on various word forms
test_words = ["running", "runner", "runs", "ran",
              "happily", "happiness", "happy",
              "connection", "connected", "connecting"]

print(f"{'Original':<15} {'Stemmed':<15}")
print("-" * 30)
for word in test_words:
    print(f"{word:<15} {stemmer.stem(word):<15}")
```

### 2.2.3 Lemmatization

Lemmatization is more sophisticated than stemming. It reduces words to their **dictionary form** (lemma) while considering the context and part of speech.

| Feature | Stemming | Lemmatization |
|---|---|---|
| Approach | Chops off affixes | Uses dictionary + morphological analysis |
| Output | May not be a real word | Always a valid word |
| Context | Ignores context | Considers part of speech |
| Example | "better" -> "better" | "better" -> "good" |
| Speed | Faster | Slower |

```python
from nltk.stem import WordNetLemmatizer

# Sample text
text = "Natural Language Processing enables computers to understand human language."

# Tokenize the text
tokens = text.split()

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the tokens
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

print("Original Tokens:")
print(tokens)
print("\nLemmatized Tokens:")
print(lemmatized_tokens)
```

```python
# Demonstrate the difference between stemming and lemmatization
test_words = ["running", "better", "computers", "studies", "geese", "wolves"]

print(f"{'Original':<15} {'Stemmed':<15} {'Lemmatized':<15}")
print("-" * 45)
for word in test_words:
    print(f"{word:<15} {stemmer.stem(word):<15} {lemmatizer.lemmatize(word):<15}")
```

### 2.2.4 Combining Text Cleaning Techniques

A robust text cleaning pipeline combines lowercasing, punctuation removal, stop word removal, stemming, and lemmatization.

```python
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Sample text
text = "Natural Language Processing enables computers to understand human language."

# Step 1: Convert to lowercase
text = text.lower()

# Step 2: Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Step 3: Tokenize the text
tokens = text.split()

# Step 4: Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Step 5: Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Step 6: Stem and lemmatize the filtered tokens
processed_tokens = [lemmatizer.lemmatize(stemmer.stem(word)) for word in filtered_tokens]

print("Original Text:")
print(text)
print("\nFiltered Tokens (Stop Words Removed):")
print(filtered_tokens)
print("\nProcessed Tokens (Stemmed and Lemmatized):")
print(processed_tokens)
```

## 2.3 Regular Expressions

Regular expressions (regex) are powerful tools for searching, matching, and manipulating text based on specific patterns. Python provides regex support through the built-in `re` module.

### 2.3.1 Basics of Regular Expressions

A regular expression is a sequence of characters that defines a search pattern. Key functions in Python's `re` module:
- `re.search()` -- Find the first match
- `re.match()` -- Match at the beginning of a string
- `re.findall()` -- Find all matches
- `re.sub()` -- Search and replace

```python
import re

# Sample text
text = "The quick brown fox jumps over the lazy dog."

# Define a pattern to search for the word "fox"
pattern = r"fox"

# Use re.search() to find the pattern in the text
match = re.search(pattern, text)

# Display the match
if match:
    print("Match found:", match.group())
    print("Position:", match.start(), "-", match.end())
else:
    print("No match found.")
```

### 2.3.2 Common Regex Patterns and Syntax

| Metacharacter | Meaning | Example |
|---|---|---|
| `.` | Any single character (except newline) | `a.b` matches "acb", "a1b" |
| `^` | Start of string | `^Hello` matches "Hello world" |
| `$` | End of string | `world$` matches "Hello world" |
| `*` | Zero or more repetitions | `ab*` matches "a", "ab", "abb" |
| `+` | One or more repetitions | `ab+` matches "ab", "abb" (not "a") |
| `?` | Zero or one repetition | `ab?` matches "a", "ab" |
| `[]` | Character set | `[aeiou]` matches any vowel |
| `\d` | Any digit `[0-9]` | `\d+` matches "123" |
| `\w` | Any word character `[a-zA-Z0-9_]` | `\w+` matches "hello_1" |
| `\s` | Any whitespace | `\s+` matches spaces/tabs |
| `\|` | OR operator | `cat\|dog` matches either |
| `()` | Grouping | `(ab)+` matches "abab" |

### 2.3.3 Practical Examples of Regex in Python

#### Example 1: Extracting Email Addresses

```python
import re

# Sample text with email addresses
text = "Please contact us at support@example.com or sales@example.com for further information."

# Define a regex pattern to match email addresses
pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b"

# Use re.findall() to find all matches
emails = re.findall(pattern, text)

# Display the extracted email addresses
print("Extracted Email Addresses:")
print(emails)
```

#### Example 2: Validating Phone Numbers

```python
import re

# Sample text with phone numbers
text = "Contact us at (123) 456-7890 or (987) 654-3210."

# Define a regex pattern to match phone numbers in (XXX) XXX-XXXX format
pattern = r"\(\d{3}\) \d{3}-\d{4}"

# Use re.findall() to find all matches
phone_numbers = re.findall(pattern, text)

# Display the extracted phone numbers
print("Extracted Phone Numbers:")
print(phone_numbers)
```

#### Example 3: Replacing Substrings

```python
import re

# Sample text
text = "The quick brown fox jumps over the lazy dog. The fox is clever."

# Define a pattern to match the word "fox"
pattern = r"fox"

# Use re.sub() to replace "fox" with "cat"
new_text = re.sub(pattern, "cat", text)

# Display the modified text
print("Original Text:")
print(text)
print("\nModified Text:")
print(new_text)
```

### 2.3.4 Advanced Regex Techniques

#### Example 4: Extracting Dates in Multiple Formats

```python
import re

# Sample text with dates in different formats
text = "The event is scheduled for 2022-08-15. Another event is on 15/08/2022."

# Define a regex pattern to match dates in YYYY-MM-DD or DD/MM/YYYY format
pattern = r"\b(?:\d{4}-\d{2}-\d{2}|\d{2}/\d{2}/\d{4})\b"

# Use re.findall() to find all matches
dates = re.findall(pattern, text)

# Display the extracted dates
print("Extracted Dates:")
print(dates)
```

#### Example 5: Extracting Hashtags from Social Media Text

```python
import re

# Sample text with hashtags
text = "Loving the new features of this product! #excited #newrelease #tech"

# Define a regex pattern to match hashtags
pattern = r"#\w+"

# Use re.findall() to find all matches
hashtags = re.findall(pattern, text)

# Display the extracted hashtags
print("Extracted Hashtags:")
print(hashtags)
```

## 2.4 Tokenization

Tokenization is a fundamental step in the text preprocessing pipeline. It involves breaking down text into smaller units called **tokens**. These tokens can be words, sentences, or even individual characters.

### Why tokenization matters:
- **Simplification**: Breaks complex text into manageable units
- **Standardization**: Creates a consistent representation of the text
- **Feature Extraction**: Facilitates extraction of meaningful features for ML models

### Types of Tokenization:
1. **Word Tokenization** -- Split text into individual words
2. **Sentence Tokenization** -- Split text into individual sentences
3. **Character Tokenization** -- Split text into individual characters

### 2.4.3 Word Tokenization

Word tokenization splits text into individual words. We demonstrate using both **NLTK** and **spaCy**.

```python
# Word Tokenization with NLTK
from nltk.tokenize import word_tokenize

# Sample text
text = "Natural Language Processing enables computers to understand human language."

# Perform word tokenization
tokens = word_tokenize(text)

print("Word Tokens (NLTK):")
print(tokens)
```

```python
# Word Tokenization with spaCy
import spacy

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Natural Language Processing enables computers to understand human language."

# Perform word tokenization
doc = nlp(text)
tokens = [token.text for token in doc]

print("Word Tokens (spaCy):")
print(tokens)
```

Notice that both NLTK and spaCy treat punctuation (the period `.`) as a separate token, unlike the naive `split()` approach.

### 2.4.4 Sentence Tokenization

Sentence tokenization splits text into individual sentences. This is useful for tasks that require sentence-level analysis, such as summarization, sentiment analysis, and machine translation.

```python
# Sentence Tokenization with NLTK
from nltk.tokenize import sent_tokenize

# Sample text
text = "Natural Language Processing enables computers to understand human language. It is a fascinating field."

# Perform sentence tokenization
sentences = sent_tokenize(text)

print("Sentences (NLTK):")
for i, sent in enumerate(sentences, 1):
    print(f"  {i}. {sent}")
```

```python
# Sentence Tokenization with spaCy
import spacy

nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Natural Language Processing enables computers to understand human language. It is a fascinating field."

# Perform sentence tokenization
doc = nlp(text)
sentences = [sent.text for sent in doc.sents]

print("Sentences (spaCy):")
for i, sent in enumerate(sentences, 1):
    print(f"  {i}. {sent}")
```

### 2.4.5 Character Tokenization

Character tokenization splits text into individual characters. It is used for tasks like text generation, handwriting recognition, spell checking, and text encryption.

```python
# Character Tokenization
text = "Natural Language Processing"

# Perform character tokenization
characters = list(text)

print("Characters:")
print(characters)
print(f"\nTotal characters: {len(characters)}")
```

### 2.4.6 Tokenization Pipeline

Let's combine different tokenization techniques into a single pipeline that demonstrates word, sentence, and character tokenization side by side using both NLTK and spaCy.

```python
import nltk
import spacy

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Natural Language Processing enables computers to understand human language. It is a fascinating field."

# 1. Word tokenization using NLTK
word_tokens = nltk.word_tokenize(text)
print("Word Tokens (NLTK):")
print(word_tokens)

# 2. Sentence tokenization using NLTK
sentence_tokens = nltk.sent_tokenize(text)
print("\nSentence Tokens (NLTK):")
print(sentence_tokens)

# 3. Sentence tokenization using SpaCy
doc = nlp(text)
spacy_sentence_tokens = [sent.text for sent in doc.sents]
print("\nSentence Tokens (spaCy):")
print(spacy_sentence_tokens)

# 4. Word tokenization using SpaCy
spacy_word_tokens = [token.text for token in doc]
print("\nWord Tokens (spaCy):")
print(spacy_word_tokens)

# 5. Character tokenization
char_tokens = list(text)
print("\nCharacter Tokens:")
print(char_tokens)
```

## Chapter 2 Summary

In this chapter we covered the foundational techniques for preparing raw text data for NLP analysis:

1. **Understanding Text Data** -- Text is inherently unstructured and requires preprocessing to handle ambiguity, variability, noise, and high dimensionality.

2. **Text Cleaning** -- Three key techniques:
   - *Stop Word Removal*: Filter out common words that carry little meaning

These preprocessing steps form the foundation of any NLP pipeline, ensuring that text data is clean, consistent, and structured for further analysis and modeling.
